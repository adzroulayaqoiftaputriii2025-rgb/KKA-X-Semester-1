#Jenis-jenis operator aritmatika pada python

```python
#Penjumlahan
iphone_rolan = 21
iphone_gibran = 4

total_iphone = iphone_rolan + iphone_gibran

print(total_iphone)
```

```python
#Pengurangan
iphone_rolan = 21
iphone_gibran = 4

selisih_iphone = iphone_rolan - iphone_gibran

print(selisih_iphone)
```

```python
#perkalian
jumlah_dus = 5
jumlah_botol_per_dus = 10

total_botol = jumlah_dus * jumlah_botol_per_dus

print(total_botol)
```

```python
#Pembagian
iphone_rolan = 21
jumlah_anak_yatim = 5

pembagian_iphone = iphone_rolan / iphone_gibran

print(pembagian_iphone)
```

```python
#Pembagian bulat
iphone_rolan = 21
jumlah_anak_yatim = 5

pembagian_iphone = iphone_rolan // iphone_gibran

print(pembagian_iphone)
```
