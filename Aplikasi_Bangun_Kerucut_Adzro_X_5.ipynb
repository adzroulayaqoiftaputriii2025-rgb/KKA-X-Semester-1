```python

#Program menghitung volume kerucut

#Meminta input dari pengguna
jari_jari_alas = float(input("Masukkan jari-jari alas kerucut(cm):"))
tinggi = float(input("Masukkan tinggi kerucut(cm):"))

#Menghitung volume
volume = (1/3) * (22/7) * jari_jari_alas**2 * tinggi

#Menampilkan hasil
print(f"Volume kerucut adalah: {volume} cm^3.")
```
